````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]
number_of_products = []

def initialize_inventory(products):
    return {product: int(input(f'How many {product} do you have in inventory? ')) for product in products}

def get_customer_orders(products, inventory):
    customer_orders = set()

    while True:
        order = input('Insert the name of a product within the list that the customer wants to order: ').lower()
        
        if order in products and inventory[order] > 0:
            customer_orders.add(order)
            add_more = input('Do you want to add another order? (YES/NO): ').upper()
            if add_more == 'NO':
                break
            elif add_more != 'YES':
                print('Invalid input. Please respond with YES or NO.')
        else:
            print('The product is not in the list or out of stock. Please try again.')
    
    return customer_orders

def update_inventory(customer_orders, inventory):
    for order in customer_orders:
        if order in inventory and inventory[order] > 0:
            inventory[order] -= 1
    return inventory

def calculate_order_statistics(customer_orders, inventory):
    total_products_ordered = len(customer_orders)
    percentage_ordered = total_products_ordered / sum(inventory.values()) * 100
    return total_products_ordered, percentage_ordered

def print_updated_inventory(inventory, products):
    in_stock_products = [(product, inventory[product]) for product in products if inventory.get(product, 0) > 0]
    for product, quantity in in_stock_products:
        print(f'Current {product} on inventory: {quantity}')

def get_prices_for_ordered_items(customer_orders):
    prices = {}
    for item in customer_orders:
        while True:
            try:
                price = float(input(f'Enter the price of {item}: '))
                if price < 0:
                    raise ValueError("Price cannot be negative.")
                prices[item] = price
                break
            except ValueError:
                print("Invalid input. Please enter a non-negative number for the price.")
    return prices

def print_total_price(customer_orders, prices):
    total_price = sum(prices[item] for item in customer_orders if item in prices)
    print(f'Total Price of the Order: ${total_price:.2f}')
````

````python
inventory = initialize_inventory(products)
customer_orders = get_customer_orders(products, inventory)
inventory = update_inventory(customer_orders,inventory)
order_statistics = calculate_order_statistics(customer_orders, inventory)
print_updated_inventory (inventory,products)
prices = get_prices_for_ordered_items(customer_orders)
print_total_price(customer_orders, prices)
````
